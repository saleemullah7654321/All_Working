```python
reader = easyocr.Reader(['en'])
```

```python
for i,j in enumerate(images):
    print(f"========== Image {i} Data  ==========")
    output = reader.readtext(rf'../NIC_Images/{j}')
    for a,b,c in output:
        print(b)
     
```
