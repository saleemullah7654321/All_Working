```python
img=np.zeros([500,500,3],dtype=np.uint8)
count=0
for i in range(1,6):
    c=np.random.randint(0,255,size=(3,))
    for j in range(count*100,i*100):
        for k in range(count*100,i*100):
            img[j][k]=c
    count=i
plt.imshow(img)
```

```python
img=np.zeros([500,500,3],dtype=np.uint8)
count=0
for i in range(1,6):
    c=np.random.randint(0,255,size=(3,))
    for j in range(count*100,i*100):
        for k in range(count*100,i*100):
            img[j][-k]=c
    count=i
plt.imshow(img)
```

```python
img=np.zeros([500,500,3],dtype=np.uint8)
count=0
for i in range(1,6):
    c=np.random.randint(0,255,size=(3,))
    for j in range(count*100,i*100):
        for k in range(count*100,i*100):
            img[j][k]=c
            img[j][-k]=c
            img[-j][k]=c
            img[-j][-k]=c
    count=i
plt.imshow(img)
```

```python
img=np.zeros([500,500,3],dtype=np.uint8)
count=0
for i in range(1,6):
    c=[255,255,255]
    for j in range(count*100,i*100):
        for k in range(count*100,i*100):
            img[j][k]=c
            img[j][-k]=c
            img[-j][k]=c
            img[-j][-k]=c
    count=i

for i in range(0,100):
    for j in range(200,300):
        img[i,j]=[255,255,255]
        img[-i,-j]=[255,255,255]
        img[j,i]=[255,255,255]
        img[-j,-i]=[255,255,255]

plt.imshow(img)
```

```python
img=np.zeros([500,500,3],dtype=np.uint8)
c=255
for i in range(0,500):
    for j in range(0,500):
        if i%100==0 and j%100==0:
            c=~c
        img[i,j]=c
plt.imshow(img)
```
