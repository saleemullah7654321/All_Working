```python
import json
import cv2
import matplotlib.pyplot as plt
```

```python
f = json.load(open('./output/my_data.json'))
f
```

```python
img=cv2.imread('../img.jpg')
img=img.copy()
plt.imshow(img)
```

```python
plt.figure(figsize=(10,10))
for i in f['Data']:
    # print(i)
    x1,y1=i['pos'][0]
    x2,y2=i['pos'][1]
    cv2.rectangle(img,(x1,y1),(x2,y2),(255,0,0),5)
    cv2.putText(img,i['text'],(x1,y1),cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,255,0),3)
plt.imshow(img)
img.shape
```
