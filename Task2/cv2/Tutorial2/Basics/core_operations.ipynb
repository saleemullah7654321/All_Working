```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('../../Tutorial1/basics/images/bird.jpg')
img=img.copy()
img.shape
```

```python
plt.imshow(img)
```

```python
b,g,r=cv2.split(img)
img=cv2.merge((b,g,r))
nimg=img[150:300,150:400]
img[0:150,0:250]=nimg
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
b,g,r=cv2.split(img)
img=cv2.merge((b,g,r))
nimg=img[150:400,150:400]
_shape=img.shape[0]-nimg.shape[0],img.shape[1]-nimg.shape[1]
img[_shape[0]:_shape[0]+nimg.shape[0],_shape[1]:_shape[1]+nimg.shape[1]]=nimg
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
b,g,r=cv2.split(img)
img=cv2.merge((b,g,r))
nimg=img[150:400,150:400]
_shape=img.shape[0]-nimg.shape[0],img.shape[1]-nimg.shape[1]
img[0:nimg.shape[1],_shape[1]:_shape[1]+nimg.shape[1]]=nimg
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
b,g,r=cv2.split(img)
img=cv2.merge((b,g,r))
nimg=img[150:400,150:400]
_shape=img.shape[0]-nimg.shape[0],img.shape[1]-nimg.shape[1]
img[_shape[1]:_shape[1]+nimg.shape[1],0:nimg.shape[1]]=nimg
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
b,g,r=cv2.split(img)
img=cv2.merge((b,g,r))
nimg=img[150:400,150:400]
nimg=cv2.resize(nimg,(img.shape[1],img.shape[0]))
img=cv2.addWeighted(nimg,.3,img,1,0)
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
