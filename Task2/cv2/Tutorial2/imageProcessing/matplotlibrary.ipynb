```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def display(*img):
    for i,j in enumerate(img):
        cv2.imshow(f'{i}',j)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
img=cv2.imread('../../Tutorial1/basics/images/flemingo.jpg',-1)
img=img.copy()
```

```python
plt.imshow(img)
plt.show()
```

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```
